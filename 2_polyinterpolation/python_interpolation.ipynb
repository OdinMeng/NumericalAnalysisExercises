```python
import numpy as np 
import numpy.linalg as LA 
import matplotlib.pyplot as plt 
```

```python
# create training dataset
train_points = 100
eps = np.random.normal(0, 1, train_points)

X = np.linspace(-5, 5, train_points)
y = (X+4) * (X+1) * (np.cos(X)-1) * (X-3) + eps

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
# Plot train and test data
fig, ax = plt.subplots()

ax.scatter(X, y, color='b', label="Train", s=10)
plt.legend()

plt.suptitle("Dataset Scatterplot")
plt.show()
```

```python
X.shape
```

```python
degree = 100
```

```python
def create_vandermonde(X, deg):
    V = np.ones((X.shape[0], deg+1))

    for i in range(1, deg+1):
        V[:, i] = (X**i).reshape(X.shape[0])

    return V
```

```python
V_ = create_vandermonde(X, degree)
```

```python
coefs = LA.solve(V_.T @ V_, V_.T @ y)
```

```python
y_hat = np.dot(V_, coefs)
```

```python
# Plot train and test data
fig, ax = plt.subplots()

ax.scatter(X, y, color='b', label="Train", s=10)
ax.scatter(X, y_hat, color='r', label='Prediction', s=10)
plt.legend()

plt.suptitle("Dataset Scatterplot")
plt.show()
```

```python
# evaluate it on more a wider range
X_2 = np.linspace(-6.5, 6.5, train_points)
V_2 = create_vandermonde(X_2, degree)
y_hat_2 = np.dot(V_2, coefs)
```

```python
# Plot train and test data
fig, ax = plt.subplots()

ax.scatter(X_2, y_hat_2, color='r', label='New data', s=10)
ax.scatter(X, y, color='b', label="Train", s=10)
plt.legend()

plt.suptitle("Dataset Scatterplot")
plt.show()
```
